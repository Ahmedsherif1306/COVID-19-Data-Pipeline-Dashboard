```python
import requests
import pandas as pd

url = "https://disease.sh/v3/covid-19/countries"

response = requests.get(url)

data = response.json()

df = pd.DataFrame(data)

print(df.head())
```

```python
from sqlalchemy import create_engine

user = "username"
password = "password"
host = "host"
port = "port"
database = "database"

engine = create_engine(
    f"postgresql://{user}:{password}@{host}:{port}/{database}"
)
```

```python
for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = df[col].apply(lambda x: str(x) if isinstance(x, dict) or isinstance(x, list) else x)
```

```python
df.to_sql(
    "covid_table",
    engine,
    if_exists="replace",
    index=False
)
```
